# Data Augumentation

Data Augumetation occurs when you create a new data based on modifications of your existing data. Data augumentation
on images would include transformations like:
    

- Flipping the image horzontally or vertically

- Rotating the image

- Zooming in or out on the image

- Cropping the image

- Varying the color on the image

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc, ndimage
import keras
from keras import backend as K
from keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
# Function for plotting images withing the jupyter notebook

def plots(ims, figsize=(12,6), rows=1, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)//rows + 1
    for i in range(len(ims)):
        sp = f.add_subplot(rows, cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15,
                        zoom_range=0.1, channel_shift_range=10, horizontal_flip=True)
```

```python
image_path = "../data/train/dog/dog.1.jpg"
```

```python
# Obtain the image

image = np.expand_dims(ndimage.imread(image_path), 0)
plt.imshow(image[0])
```

### This is the image we are going to perform data augumentation with

```python
# Generate batches of augumented images from this image

aug_iter = gen.flow(image)
```

```python
# Getting 10 samples of our augumented images

aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
# Augumented Images

plots(aug_images, figsize=(20, 7), rows=2)
```
